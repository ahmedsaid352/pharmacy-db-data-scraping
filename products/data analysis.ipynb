```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv('drugs.csv')
```

```python
df.head()
```

```python
plt.figure(figsize=(20,10))
sns.boxplot(data = df.price,orient='h');
plt.title('Price range',fontsize=18)
plt.show()
```

```python
mean_price =  df.groupby('type')['price'].mean().to_frame()
```

```python
mean_price['type'] = mean_price.index
mean_price.head()
```

```python
mean_price =  mean_price.sort_values(by = 'price', ascending=False)
plt.figure(figsize=(20,10))
sns.barplot(y=mean_price.type,x=mean_price.price,orient='h');
plt.title('categories mean prices',fontsize = 18)
plt.xticks(fontsize = 18)
plt.yticks(fontsize = 18)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.boxplot(x=df.price,y= df.type, data = df);
plt.title('price range for different types',fontsize = 18)
plt.xticks(fontsize = 18)
plt.yticks(fontsize = 18)
plt.show()
```
